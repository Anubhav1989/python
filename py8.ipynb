```python
# Data(df) -> x(independent data)+y(dependent or target data)
# x -> x_train,x_test
# y -> y_train,y_test
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
```

```python
df.shape
```

```python
x=df.drop(columns=['has_covid'])# Independent Data
x.shape
```

```python
y=df['has_covid']# Target Data
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,
                                                  test_size=0.2,
                                                  random_state=11)
```

```python
print('total shape of df:',df.shape)
print('---------------------------')
print('input data shape:',x.shape)
print('x_train data shape:',x_train.shape)
print('x_test data shape:',x_test.shape)
print('---------------------------')
print('target data shape:',y.shape)
print('y_train data shape:',y_train.shape)
print('y_test data shape:',y_test.shape)
print('---------------------------')
```

```python
df=pd.read_csv('/content/titanic - titanic.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Fare']=df['Fare'].fillna(df['Fare'].mean())
df=df.drop(columns=['Cabin'])
df.isnull().sum()
```

```python
df.shape
```

```python
x=df.drop(columns=['Survived'])
x.shape
```

```python
y=df['Survived']
y.shape
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,
                                               test_size=0.2,
                                               random_state=13)
```

```python
print('total shape of df:',df.shape)
print('---------------------------')
print('input data shape:',x.shape)
print('x_train data shape:',x_train.shape)
print('x_test data shape:',x_test.shape)
print('---------------------------')
print('target data shape:',y.shape)
print('y_train data shape:',y_train.shape)
print('y_test data shape:',y_test.shape)
print('---------------------------')
```
