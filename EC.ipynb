```python
# Encoding->transform categorical data into numerical data
#(1).Label->1-1 column apply but we cannot see the order manually
#(2).Ordinal->multi-column apply and we can set the order manually
#(3).OneHotEncoder->has_covid[yes,no]->2 column create->covid_yes,covid_no
#(4).get_dummies->has_covid[yes,no]->2 columns create->column name with values
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
```

```python
df.head(2)
```

```python
df.dropna()
```

```python
df=df.drop(columns=['city'])
```

```python
#Approach-1->Label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
# Approach-2->Ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head(2)
```

```python
df=df.drop(columns=['city','age','fever'])
```

```python
df=df.dropna()
```

```python
df.head(2)
```

```python
df_new=df.copy()
```

```python
oe=OrdinalEncoder(categories=[['Male','Female'],
                              ['Mild','Strong'],
                              ['Yes','No']])
```

```python
oe_df=oe.fit_transform(df)
```

```python
df_oe=pd.DataFrame(oe_df,columns=df_new.columns)
```

```python
df_oe
```

```python
# Approach-3->OneHotEncoder
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head(2)
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
ohe.fit_transform(df[['gender','cough','city','has_covid']])
```

```python
# Approach-4->get_dummies
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head(2)
```

```python
df=df.dropna()
```

```python
df=pd.get_dummies(df,columns=['gender','cough','city','has_covid'],drop_first=True)
```

```python
df=df.astype(int)
```
