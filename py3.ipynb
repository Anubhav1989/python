```python
import matplotlib.pyplot as plt
#here the pyplot is the sub module which is used to plot the graphs
```

```python
#Line plot
import pandas as pd
data={
    'date':[11,12,13],
    'sales':[100,200,300],
    'stock':[250,300,350]
}
df=pd.DataFrame(data)
```

```python
#line plot is used when we have to visualize the time series data
#into the time series data we should always plot the time on the x-axis
#stock->t-axis
#syntax->plt.plot(x-axis,y-axis)
plt.plot(df['date'],df['sales'],label='sales')
plt.plot(df['date'],df['stock'],label='stock')
plt.title('date vs sales')
plt.xlabel('date')
plt.ylabel('sales')
plt.legend()
plt.grid()
plt.show()#used to show the graph which is built
```

```python
#Bar plot
student_data={
    'name':['xp','ts','ac','d','e','f'],
    'marks':[20,35,22,19,13,11]
}
sdf=pd.DataFrame(student_data)
```

```python
#bar plot is used when we want to plot a graph b/w the categorical
#and numerical data to find frequency of each of the category
plt.bar(sdf['name'],sdf['marks'])#bar-when less categories
#barh->used when we have more number of categories
plt.title('student vs marks')
plt.xlabel('name')
plt.ylabel('marks')
plt.show()
```

```python
#scatter plot
data={
    'area':[200,250,275,300],
    'price':[11,13,19,22],
    'price1':[50,40,30,20]
}
hdf=pd.DataFrame(data)
```

```python
#scatter plot is used to find the relationship b/w the two features
#area->price->direct relationship->area inc.->price inc.
#area->price1->indirect relationship->area inc.->price dec.
plt.scatter(hdf['area'],hdf['price1'])
```
